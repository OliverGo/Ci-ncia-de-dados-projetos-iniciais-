Qual técnica de aprendizado não supervisionado você recomendaria para reduzir a dimensionalidade do conjunto de dados e qual é o objetivo principal dessa técnica?
 - PCA, diminuir a dimensonalidade dos dados matendo a parte importante. A tecnica foi escolhida pois reduz a dimensonalidade dos dados, facilitando seu processamento e evitando valores absurdos 

Após a redução da dimensionalidade, a empresa deseja encontrar quais produtos são frequentemente comprados juntos. Qual método de associação seria mais adequado para essa tarefa e por quê?

- Apriori para usar regra de associação, para esse problema séra implementado para associar produtos que são comprados juntos com frequência 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

 
np.random.seed(42)
df = pd.DataFrame({
    'idade': np.random.randint(18, 70, 100),
    'frequencia_compras': np.random.randint(1, 30, 100),
    'valor_medio_gasto': np.random.randint(10, 500, 100)
})

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

df_pca = pd.DataFrame(df_pca, columns=['Componente 1', 'Componente 2'])

plt.figure(figsize=(8, 6))
plt.scatter(df_pca['Componente 1'], df_pca['Componente 2'], alpha=0.7)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Redução de Dimensionalidade com PCA')
plt.show()

# Um exemplo simples de como usar PCA para reduzir a dimensionalidade dos dados, usando o matplotlib para visualizar os dados
```
